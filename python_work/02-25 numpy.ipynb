```python
a = np.arange(1, 10).reshape(3, 3)
print(a)
print(a > 5)    # a 요소 하나하나 > 5인지 불리언값 하나하나 반환
print(np.all(a > 5))    # a 요소 모두가 > 5인지 불리언값 한개만 반환
print(np.any(a > 5))    # a 요소 중 하나라도 > 5인지 불리언값 한개만 반환
```

```python
# 행렬의 복사
a = np.arange(1, 10).reshape(3, 3)
b = a   # 얕은 복사: 참조변수 a에 저장된 행렬의 주소를 참조변수 b한테도 주는 거임
b[0, 0] = 10
print(a)
print(b)    # 다 아는 결과

c = a.copy()    # 깊은 복사: a와 똑같은 행렬을 하나 더 만들어서 c에 집어넣은 거임
print(c)
c[0, 0] = 20
print(a)
print(c)    # c의 값만 바뀌어있는 것을 볼 수 있음

print(id(a))
print(id(b))    # 얕은 복사는 주소가 같은데
print(id(c))    # 깊은 복사는 주소 자체가 다름 ㅇㅇ
```

```python
# 행렬 정렬!!!!!!!
arr = np.array([[3, 2, 4], [2, 7, 6], [9, 5, 1]])
print(arr)
print(np.sort(arr)) # 행마다 오름차순으로 정렬
print(np.sort(arr, axis=1)) # 행마다 오름차순으로 정렬, 기본값과 같음
print(np.sort(arr, axis=0)) # 열마다 오름차순으로 정렬
print(np.sort(arr, axis=0)[::-1]) # 열마다 오름차순으로 정렬인데, 행 순서를 역순으로 정렬
print(np.sort(arr, axis=0)[:, ::-1]) # 열마다 오름차순으로 정렬인데, 열 순서를 역순으로 정렬
print(np.sort(arr, axis=0)[::-1, ::-1]) # 열마다 오름차순으로 정렬인데, 행이랑 열을 전부 역순으로 정렬
```

```python
data = np.loadtxt('data/chocolate_rating.csv', delimiter=',')   # ','를 구분자로 csv를 행렬로 읽어봐버리기
print(data)
print()
print('차원의 수: ', data.ndim)
print('모양: ', data.shape)
print('원소의 수: ', data.size)
print('원소의 크기: ', data.itemsize)
print('행렬의 용량: ', data.nbytes)

ratings_mean = data[:, 3].mean()
print('모든 초콜릿의 평점: ', ratings_mean, '점')

high_level = data[data[:, 3] >= 4]
print('우수 초콜릿의 수: ', high_level.shape[0])

high_id = high_level[:, 0].astype(np.int64)
print('우수 초콜릿의 id: ', high_id)

high_kakao = high_level[:, 2]   # 우수 초콜릿의 카카오 함유량
unique_values, value_count = np.unique(high_kakao, return_counts=True)  # 중복값 제거 함유량값, 빈도수
t = list(zip(unique_values, value_count))
for k, v in t:
    print(k, v)
```

```python
# 카카오 함유량 빈도수가 가장 높은 함유량
max_index = np.argmax(value_count)  # np.argmax() = 배열에서 가장 큰 값이 있는 위치(index) 리턴
print(value_count)
print(max_index)
max_kakao = unique_values[max_index]
print(f'우수 초콜릿 {len(high_id)}가지 중에서 카카오 함유량 {max_kakao}%가 빈도수 {value_count[max_index]}로 가장 높습니다.')
```

# 영화 평점 데이터 실습
<h3>
    1. 사용자별 영화 평점의 평균을 구하라<br>
    2. 영화별 영화의 평점의 평균을 구하라.<br>
</h3>

```python
# 데이터 가져오기
data = np.genfromtxt('data/ratings.dat', delimiter='::', dtype=np.int64)    #genfromtxt는 loadtxt와는 달리 구분자를 여러 문자로 쓸수 있음
```

```python
# 1. 사용자별 영화 평점의 평균
user_ids = np.unique(data[:, 0])

mean_values = []
for user_id in user_ids:
    user_datas = data[data[:, 0] == user_id, :]
    mean_value = user_datas[:, 2].mean()
    mean_values.append([user_id, mean_value])

means_array = np.array(mean_values)

np.savetxt('data/result.csv', means_array, fmt='%.1f', delimiter=',')
```

```python
# 2. 영화별 영화의 평점 평균
movie_ids = np.unique(data[:, 1])

means_movie = []

for movie_id in movie_ids:
    movie_data = data[data[:, 1] == movie_id, :]
    mean_movie = movie_data[:, 2].mean()
    means_movie.append([movie_id, mean_movie])

means_movie_array = np.array(means_movie)

np.savetxt('data/means_movie.csv', means_movie_array, fmt='%s', delimiter='::')
```
