```python
import numpy as np
import pandas as pd
```

```python
# 세팅
name = ['허준호', '이가원', '배규민', '고고림', '이새봄', '이보람', '이루리', '오다현']
gender = ['남자', '여자', '남자', '남자', '여자', '여자', '여자', '여자']
age = [30, 24, 23, 21, 28, 26, 24, 24]
height = [183, 162, 179, 182, 160, 163, 157, 172]

data = {'이름': name, '성별': gender, '나이': age, '키': height}
df = pd.DataFrame(data)
df.나이.astype(np.int64)
df.키.astype(np.int64)

df
```

```python
# 1. 몸무게 column 추가
df['몸무게'] = [82.5, 50.0, 78.0, 80.5, 48.0, 51.2, 46.0, 65.0]
df.몸무게.astype(np.float64)
df
```

```python
# 2. 키가 180 이상인 사람의 이름과 키를 조회하라
df.loc[df['키'] >= 180, ['이름', '키']]
```

```python
# 3. 여자 중 키가 160보다 큰 사람을 조회하라.
df.loc[(df['성별'] == '여자') & (df['키'] >= 160)]
```

```python
# 4. 성별별 키의 평균을 구하라
df.groupby('성별')['키'].mean()
```

```python
# 5. 성별별 키와 몸무게의 평균, 합, 인원수를 구하라
df.groupby('성별')[['키', '몸무게']].agg(['mean', 'sum', 'count'])
```

```python
# 6. 이름과 성별 칼럼을 추출하여 df2를 생성하라
df2 = df.loc[:, ['이름', '성별']]
df2
```

```python
# 7. df2에 [연락처] 칼럼을 추가한다. (데이터: ['010-1111-2525', ... 등 8개])
df2['연락처'] = ['010-1111-1111', '010-2222-2222', '010-3333-3333', '010-4444-4444', '010-5555-5555', '010-6666-6666', '010-7777-7777', '010-8888-8888']
df2
```

```python
# 8. df와 df2의 이름 칼럼을 인덱스로 설정한다.
# df.set_index('이름', inplace=True)
# df2.set_index('이름', inplace=True)
print(df)
print(df2)
```

```python
# 9. df와 df2를 인덱스를 기준으로 조인한다.
join_df = df.merge(df2, left_index=True, right_index=True)
join_df
```
