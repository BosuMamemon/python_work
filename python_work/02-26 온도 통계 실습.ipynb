```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/temperature_2014.csv')
data.info()
```

```python
# 날짜 자료형 바꾸기
data['PST'] = pd.to_datetime(data['PST'])

data.info()
```

```python
# 한글 폰트 설정
import matplotlib.font_manager as fm

plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
# 기본 설정
fig = plt.Figure(dpi=128, figsize=(10, 6))
# 각 행을 미리 리스트로 변경(시리즈 바뀌는 걸 방지하기 위함)
pst = list(data['PST'])
high = list(data['Max TemperatureF'])
low = list(data['Min TemperatureF'])
```

```python
# 차트 그리기
plt.plot(pst, low, c='blue', alpha=0.75)
plt.plot(pst, high, c='red', alpha=0.75)
plt.title('온도 통계', fontsize=18)
plt.xlabel('기간', fontsize=14)
plt.ylabel('온도', fontsize=14)
plt.grid()
plt.fill_between(pst, high, low, facecolor='skyblue', alpha=0.5)    # pst를 색칠하는데 high랑 low 사이를 칠하겠다는 뜻임
plt.savefig('data/chart_test.png')
plt.show()
```
