```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
# matplotlib.pyplot으로 plot 차트 그리기
def chart1():
    x = [1, 2, 3, 4, 5]
    y = [10, 20, 25, 30, 40]

    fig = plt.figure(figsize=(4, 3), dpi=128)   # 차트의 사이즈를 (4, 3)사이즈에 해상도는 128도트로
    
    plt.plot(x, y)  # 차트 그리기
    plt.title('Simple Dimple')  # 차트 이름 붙이기
    plt.xlabel('Simple')
    plt.ylabel('Dimple')    # x, y축 제목 붙이기
    plt.savefig('data/chart1.png')  # 무조건 show()보다 앞에 해야함. show()를 하면 차트가 사라짐...
    plt.show()

chart1()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [25, 25, 25, 25, 25]
y2 = [0, 25, 50, 25, 0]

plt.plot(x, y1, label='Red', color='blue', linestyle='--', marker='o')
plt.plot(x, y2, label='Blue', color='red', linestyle='-', marker='s')
plt.title('A')
plt.xlabel('WHAT')
plt.ylabel('FUCK')
plt.legend()    # 범례를 뜻함
plt.grid(True)  # 그리드 배경 True
plt.savefig('data/chart2.png')
plt.show()
```

```python
# bar 차트 그리기
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.bar(categories, values, color=['blue', 'red', 'green', 'purple'])
```

```python
# bar 차트 그리기(2)
categories = ['A', 'B', 'C', 'D']
values1 = [10, 20, 15, 25]
values2 = [15, 12, 20, 15]
plt.bar(categories, values1, color=['blue', 'red', 'green', 'purple'])
plt.bar(categories, values2, color=['pink', 'brown', 'violet', 'red'], bottom=values1)  # values1값을 이 바의 아래에 두겠다는 말임
```

```python
# Histogram 차트

data = np.random.randn(1000)    # 정규 분포를 따른 난수 1000개 생성

plt.hist(data, bins=25, color='pink', alpha=0.8)    # bins는 막대 개수, alpha는 불투명도
plt.title('PINK')
plt.xlabel('Tree')
plt.ylabel('So pretty')
plt.savefig('data/chart3.png')
plt.show()
```

```python
# scatter 그래프(산점도 그래프)
x = np.random.rand(25)
y = np.random.rand(25)

colors = np.random.rand(25)
sizes = np.random.rand(25) * 4000

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='plasma') # viridis는 색맹 친화적 컬러맵을 쓴다는 뜻
plt.title('WOW')
plt.xlabel('BUBBLE BUBBLE')
plt.ylabel('I LOVE PINK')
plt.colorbar()
plt.savefig('data/chart4.png')
plt.show()
```

```python
# pie형 차트 만들기!!
labels = ['A', 'B', 'C', 'D']
sizes = [10, 20, 30, 40]
colors = ['brown', 'yellow', 'red', 'green']
explode = (0, 0.2, 0, 0)    # 두번째 조각 강조

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', explode=explode, startangle=80)
plt.title('PIZZA')
plt.savefig('data/chart5.png')
plt.show()
```

```python
# 서브플롯 설정법
# + 한글 설정법
import matplotlib.font_manager as fm

plt.rc('font', family='Malgun Gothic')  # 차트에 한글 설정법
plt.rc('axes', unicode_minus=False) # 차트에 한글 설정하고 나서 마이너스 기호 표시하려면 이거 무조건 해줘야됨
# 결국 이렇게 한 세트임

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.Figure(figsize=(10, 5)) # 가로 10인치 세로 5인치라는 뜻

plt.subplot(2, 1, 1)    # 2행 1열의 서브플롯의 1번째라는 뜻
plt.plot(x, y1, label='흔들흔들 열매', color='skyblue')
plt.title('그라라라라라')
plt.legend()

plt.subplot(2, 1, 2)    # 2행 1열의 서브플롯의 2번째라는 뜻
plt.plot(x, y2, label='어둠어둠 열매', color='grey')
plt.title('제하하하하하')
plt.legend()
plt.tight_layout()
plt.savefig('data/chart6.png')
plt.show()
```
